# Skills Network Diagram

```python
import numpy as np
import pandas as pd 
import networkx as nx
from pyvis.network import Network

# Load data from Excel
relationships_df = pd.read_excel('skill_node_attributes.xlsx', sheet_name='relationships')
nodes_df = pd.read_excel('skill_node_attributes.xlsx', sheet_name='nodes_attributes')

# Create graph from relationships
G = nx.Graph()
G = nx.from_pandas_edgelist(relationships_df, source='Skill', target='Successor')

# Initialize pyvis network
net = Network(notebook=True, cdn_resources='in_line', select_menu=True)

# Add nodes with attributes from nodes_df
for node in G.nodes():
    # Fetch the corresponding row in nodes_df for node attributes
    node_data = nodes_df[nodes_df['skill_node'] == node]
    
    if not node_data.empty:
        size = int(node_data['size'].values[0])  # Convert to standard int
        color = str(node_data['color'].values[0])  # Ensure it's a string
        net.add_node(node, title=node, size=size, color=color)
    else:
        # Default size and color if not in nodes_df
        net.add_node(node, title=node, size=10, color='#97C2FC')

# Add edges from G to pyvis network
for edge in G.edges():
    net.add_edge(edge[0], edge[1])

# Custom options (physics settings)
net.set_options(
    """
    const options = {
        "physics": {
            "maxVelocity": 1,
            "minVelocity": 0.75
        }
    }
    """
)

# Generate the HTML file
net.show('skills_diagram2.html')
```
