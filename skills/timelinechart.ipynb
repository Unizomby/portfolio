```python
import plotly.express as px
import pandas as pd

# Create a sample DataFrame with task information
df = pd.DataFrame([
    dict(Task="Operations Eng Asc", Start='2009-01-01', Finish='2012-08-01', Experience="Work"),
    dict(Task="Industrial Eng Asc", Start='2012-08-01', Finish='2014-03-01', Experience="Work"),
    dict(Task="Industrial Eng", Start='2014-03-01', Finish='2015-08-01', Experience="Work"),
    dict(Task="Industrial Eng Sr", Start='2015-08-01', Finish='2017-10-01', Experience="Work"),
    dict(Task="Industrial Eng Stf", Start='2017-10-01', Finish='2019-10-01', Experience="Work"),
    dict(Task="Industrial Eng Asc Mgr", Start='2019-10-01', Finish='2022-10-01', Experience="Work"),
    dict(Task="Industrial Eng Mgr", Start='2022-10-01', Finish='2024-01-01', Experience="Work"),
    dict(Task="IE BS - UTEP", Start='2008-10-01', Finish='2008-12-31', Experience="Education"),
    dict(Task="IE MS - NMSU", Start='2010-09-01', Finish='2012-05-01', Experience="Education")
])

# Create the timeline chart
fig = px.timeline(df, x_start="Start", x_end="Finish", y="Experience", color="Task", color_discrete_sequence=px.colors.sequential.Purp)
fig.update_layout(showlegend=False, # Hide the legend
                  plot_bgcolor="#fbfbfe",
                  paper_bgcolor='#fbfbfe',
                  font=dict(family="Arial", size=14, color="RebeccaPurple"),
                  yaxis_tickangle=-90,      
                  )
fig.update_traces(width=0.3)

# Add annotations (labels) for each task using arrows
for index, row in df.iterrows():
    fig.add_annotation(
        x=row['Start'],
        y=row['Experience'],
        text=row['Task'],
        showarrow=True,
        arrowhead=1,
        ax=0,
        standoff=20,
        font=dict(family="Arial", size=8, color="RebeccaPurple"),
        ay=-60 # Adjust the arrow position

    )

fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Create a sample DataFrame with task information
df = pd.DataFrame([
    dict(Task="Job A", Start='2009-01-01', Finish='2009-02-28', Resource="Alex"),
    dict(Task="Job B", Start='2009-03-05', Finish='2009-04-15', Resource="Alex"),
    dict(Task="Job C", Start='2009-02-20', Finish='2009-05-30', Resource="Max")
])

# Create the timeline chart with tasks on the same line and different colors
fig = px.timeline(df, x_start="Start", x_end="Finish", y="Resource", color="Resource")
fig.update_traces(width=0.05)
fig.update_layout(plot_bgcolor='black', paper_bgcolor='black')  # Set the background color to black
fig.show()
```
